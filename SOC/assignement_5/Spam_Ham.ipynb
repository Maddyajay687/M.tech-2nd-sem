```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import string
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('spam.csv',encoding = "latin")
```

```python
data = data.dropna(axis=1)
```

```python
data.head()
```

```python
def text_process(mess):
    nopunc =[char for char in mess if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
import nltk
nltk.download('stopwords')
```

```python
data_list = data['Message'].apply(text_process)
```

```python
X = data.iloc[:,-1]
Y = data.iloc[:,0]
```

```python
X_train = X.loc[0:int(0.7 * len(X)),]
Y_train = Y.loc[0:int(0.7 * len(Y)),]

X_test = X.loc[int(0.7 * len(X)):,]
Y_tets = Y.loc[int(0.7 * len(Y)):,]
```

```python
P_spam  = np.sum(Y_train == 'spam')/ len(Y_train)
P_ham = np.sum(Y_train == 'ham') / len(Y_train)
```

```python
spam_dict = {}
ham_dict = {}
def freq(X,Y):
    for i in range(X.shape[0]):
        if(Y[i] == 'spam'):
            #print(X[i])
            text_processed = text_process(X[i])
            for data in text_processed:
                spam_dict[data] = spam_dict.get(data,0) + 1
        if(Y[i] == 'ham'):
            text_processed = text_process(X[i])
            for data in text_processed:
                ham_dict[data] = ham_dict.get(data,0) + 1
    return spam_dict,ham_dict
```

```python
spam_freq, ham_freq = freq(X_train, Y_train)
```

```python
spam_words = sum(spam_freq.values())
ham_words = sum(ham_freq.values())
```

```python
def find_prob(X, Y):
    prob_spam = {}
    prob_ham = {}
    for i in X:
        prob_spam[i] = (spam_dict[i] + 1)/ (spam_words + len(list(spam_dict.keys())))
    for i in Y:
        prob_ham[i] = (ham_dict[i] + 1)/ (ham_words + len(list(ham_dict.keys())))
        
    return prob_spam,prob_ham
```

```python
def train(X):
    result = []
    spam_prob,ham_prob = find_prob(spam_freq,ham_freq)
    for i in range(X.shape[0]):
        print(X[i])
        text_processed = text_process(X[i])
        p_spam = 0
        p_ham = 0
        for word in text_processed:
            if word in spam_prob:
                p_spam += np.log(spam_prob[word])
            else:
                p_spam -= np.log(spam_words + len(list(spam_freq.keys())))
        for word in text_processed:
            if word in ham_prob:
                p_ham += np.log(ham_prob[word])
            else:
                p_ham -= np.log(ham_words + len(list(ham_freq.keys())))
        p_spam += np.log(P_spam)
        p_ham += np.log(P_ham)
        if (p_spam >= p_ham):
            result.append('spam')
        else:
            result.append('ham')
    return result
```

```python
pred_label = train(X_train)
```

```python
df = pd.DataFrame(data = {'Actual':Y_train,'Predicted':pred_label})
```

```python
def confusion_matrix(X,Y):
    matrix = []
    fp = 0
    fn = 0
    tp = 0
    tn = 0
    for i in range(X.shape[0]):
        if(X[i] == Y[i]):
            #print(X[i])
            if(X[i] == 'spam'):
                tp += 1
            if(X[i] == 'ham'):
                tn += 1
        else:
            #print(X[i])
            if(X[i] == 'spam'):
                fn += 1
            if(X[i] == 'ham'):
                fp += 1
    matrix.append([tp,fp])
    matrix.append([fn,tn])
    print("Confusion Matrix is: ")
    for data in matrix:
        print(data)
```

```python
def accuracy(df):
      return np.sum(df['Actual'] == df['Predicted'])/df.shape[0]
```
