# Adding a bias term in the independent variable X


```python
X = np.c_[np.ones(X.shape[0]),X]
```

```python
theta = np.matmul(np.linalg.inv(np.matmul(X.T,X)),np.matmul(X.T,y))
print(theta)
```

```python
def gradient_descent(x,y,theta,num_iterations,alpha):
    past_costs = []
    past_thetas = []
    for i in range(num_iterations):
        prediction = np.dot(x,theta)
        error = prediction - y
        cost = 1 / (2 * m) * np.dot(error.T, error)
        past_costs.append(cost)
        theta = theta - (alpha * (1/m) * np.dot(x.T, error))
        past_thetas.append(theta)
    return past_thetas, past_costs
```

```python
alpha = 0.01
iterations = 4000
m = y.size
np.random.seed(123)
theta1 = np.random.rand(11)
```

```python
weight,cost = gradient_descent(X,y,theta1,iterations,alpha)
z = weight[-1]
print(z)
```

```python
plt.title('Cost Function J(Theta)')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.plot(cost)
plt.show()
```

```python
predicted_val = []
for data in X:
    predicted_val.append(np.dot(data,z))
```
